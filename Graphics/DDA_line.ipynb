DDA Line Drawing Algo

Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as mp
```

Function

```python
def DDA(x1,y1,x2,y2):
    dx = x2-x1
    dy = y2-y1
    if(abs(dx)>abs(dy)):
        steps = dx
    else:
        steps = dy
    x_inc = dx/steps
    y_inc = dy/ steps
    X = []; Y = []
    for i in range(steps):
        X.append(x1)
        Y.append(y1)
        x1 = round(x1 + x_inc)
        y1 = round(y1 + y_inc)
    return X,Y
```

Example and Plot

```python
X, Y = DDA(0,0,23,25)
```

```python
mp.scatter(X,Y)
mp.show()
```
