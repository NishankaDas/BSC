Bresenham Line Algorithm

*Importing Libraries*

```python
import matplotlib.pyplot as mp
import numpy as np
```

*Function*

```python
def line(x1, y1, x2, y2):
    x = x1
    y = y1
    dx = x2-x1
    dy = y2-y1
    P = 2*dx - dy
    Points = []
    X = []
    Y = []
    while(x<= x2):
        newPoints = [x,y]
        X.append(x)
        Y.append(y)
        Points.append(newPoints)
        x += 1
        if(P < 0):
            P = P +2*dy
        else:
            P = P + 2*dy - 2*dx
            y +=1
    return Points, X,Y
```

```python
List, X, Y =  line(0,0,250,450)
```

*Plot*

```python
mp.plot(List)
mp.scatter(X,Y)
mp.show()
```
